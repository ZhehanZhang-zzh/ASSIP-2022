```python
import json
import pandas as pd
pd.__version__
```

```python
videos = {}

with open('../data_csv/072622_samples/beauty_channel_videos.json') as json_file:
    data = json.load(json_file)
    i=0
    
    for key in data:
        print(key)
        
        for value in data[key]:
            for v in value:
                videos[i] = v
                i=i+1
```

```python
df = pd.json_normalize(videos.values())
```
